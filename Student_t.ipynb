### 7.3.3 スチューデントのｔ分布

```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import t as student_t
from scipy.stats import t
from scipy.stats import norm
```

```python
# Define the distribution parameters to be plotted
mu = 0
k_values = [100, 30, 5, 1]  # degree of freedom (= n-1)
x = np.linspace(-10, 10, 1000)
```

```python
# studentのt分布の表示 (Fig. 6.6)
for k in k_values:
    dist = student_t(k)  # 配列化
    label = r'$\mathrm{t} (df=%.0f)$' % k
    plt.plot(x, dist.pdf(x), label=label)

plt.xlim(-5, 5)
plt.ylim(0.0, 0.45)
plt.xlabel('$x$')
plt.ylabel('pdf')
plt.title("$t$ -distribution (df: degree of freedom)")

# 正規分布の表示
plt.plot(x,norm.pdf(x),lw=1,ls="--", color="r")
plt.legend();
```

### t 分布の確率密度関数とalpha 値

```python
# 標本サイズと自由度の設定
sample = 11       # 標本数 n
df = sample - 1   # 自由度 n-1

# 積分上限値の設定
x_cutoff = 1.

x1 = np.linspace(-5, 5, 100)
plt.plot(x1, t.pdf(x1, df), color="r", alpha=0.6, label="Student_t")

x2 = np.linspace(x_cutoff, 5, 100)
plt.plot(x2, t.pdf(x2, df), color="r", alpha=0)
plt.fill_between(x2, 0, t.pdf(x2, df), facecolor='m', alpha=0.3)
plt.vlines(x=x_cutoff, ymin=0, ymax=t.pdf(x_cutoff, df), linestyles='dashed', linewidths=1)
plt.legend();

print(f"下側面積 \t {t.cdf(x_cutoff, df):2.4f}")
print(f"上側面積 \t {1-t.cdf(x_cutoff, df):2.4f}  *red area*")
```

```python
# Student's t分布表の生成（表7.2）
# 自由度(degree of freedom: df)の設定
df = 5
alpha = [0.1, 0.05, 0.025, 0.01, 0.005, 0.0025]

print("【degree of freedom】:", df)
for k in alpha:
    print(f"alpha: {k:6.2%}  {t.isf(k, df, loc=0, scale=1):6.3f}")
```
