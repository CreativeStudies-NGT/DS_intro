### 誕生日のパラドックス（問題2-7）

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
days = 365  # days per year
member = 23 # number of people in a class

np.arange(days, days - member, -1)
```

```python
sol = 1 - np.prod(np.arange(days, days - member, -1) / days)
print(sol)
```

```python
# 関数を使用
def birthday(size):
    days = 365
    return 1 - np.prod(np.arange(days, days - size, -1) / days)

birthday(23)
```

```python
x = np.arange(1,71)
prob = []
for i in x:
    prob.append(birthday(i))
y=prob
plt.plot(x, y,'c')
plt.grid(color='gray',linestyle=':')
```
