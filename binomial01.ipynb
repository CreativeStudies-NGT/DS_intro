### 4.1.3 二項分布の形

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
n=15     # 試行回数
p=0.5    # 成功確率

x = np.arange(0,n+1)  # 成功回数のリスト
print(x)

y = binom.pmf(x,n,p)  # 二項分布の確率質量関数 (PMF)
print(y)
```

```python
plt.bar(x, y)
plt.grid(color='gray',linestyle=':')
```

```python
# PMFの規格化を確認
print(f"sum of all bars:{np.sum(y): 1.3f}")
```

### 成功確率 *p* = constで、試行回数（n） を変えた場合の二項分布

```python
p = 0.5                    # 成功確率
n = [5, 15, 30]            # 試行回数(list)
x = np.arange(0,max(n)+1)  # 成功回数(list)

for k in n:
    y = binom.pmf(x,k,p)
    plt.bar(x, y, alpha=0.6)
plt.grid(color='gray',linestyle=':');
```

### 試行回数固定(n=50)で、成功確率(*p*)を変えた場合の二項分布

```python
n=50    # 試行回数
p = [0.05, 0.1, 0.25, 0.5, 0.7]  # 成功確率(list)
x = np.arange(0,n+1)             # 成功回数(list)

for k in p:
    y = binom.pmf(x,n,k)
    plt.bar(x, y, alpha=0.6)
plt.grid(color='gray',linestyle=':')
```
