## 5.2 二項分布から正規分布へ

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

p = 0.4    # 成功確率
ns = [4,12,60,120]
x = np.arange(0,max(ns)+1)  # 成功回数(list)
prob = []
mu = []; var = []

for n in ns:
    prob.append(binom.pmf(x,n,p))
    mu.append(n*p)
    var.append(n*p*(1-p))
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(7,4.5))
plt.subplots_adjust(wspace=0.35, hspace=0.5)

axes[0,0].bar(x, prob[0], linewidth=2)
axes[0,0].set_title('p=0.4 / n=4')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('probability')
axes[0,0].set_xlim(-1, 6)
axes[0,0].grid(True)

axes[0,1].bar(x, prob[1], linewidth=2)
axes[0,1].set_title('p=0.4 / n=12')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('probability')
axes[0,1].set_xlim(-1, 10)
axes[0,1].grid(True)

axes[1,0].bar(x, prob[2], linewidth=2)
axes[1,0].set_title('p=0.4 / n=60')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('probability')
axes[1,0].set_xlim(0, 40)
axes[1,0].grid(True)

axes[1,1].bar(x, prob[3], linewidth=2)
axes[1,1].set_title('p=0.4 / n=120')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('probability')
axes[1,1].set_xlim(0, 80)
axes[1,1].grid(True)
```

## 同一の $\mu$, $\sigma$ を持つ二項分布と正規分布の比較
二項分布 ${}_n C _x p^x(1-q)^{n-x}$に対する、$n \to \infty $ なる連続近似に対応<p>
平均： $\mu=np$、標準偏差： $\sigma=\sqrt{np(1-p)}$の正規分布と比較してみる

```python
p = 0.4     # 成功確率
n = 100     # 試行回数
mu = n*p
std = np.sqrt(n*p*(1-p))
x = np.arange(0, n+1)  # 成功回数(list)
bi_dist = binom.pmf(x,n,p)
bi_dist[:10]
```

```python
from scipy.stats import norm
plt.bar(x, bi_dist, color="c")  # 二項分布

x_norm = np.linspace(0,80,128)
plt.plot(x_norm, norm.pdf(x_norm, loc=mu, scale=std))  # 正規分布
plt.xlim(0, 80)
print("平均: ", mu, "  分散: ", std)
```
