```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 平均を指定
mean = [60, 60]
# 共分散行列を指定
s_x = 60; s_y = 60
s_xy = 50
cov = s_xy/np.sqrt(s_x*s_y)
cov_mat = np.array([[s_x, s_xy],[s_xy, s_y]])

x, y = np.random.multivariate_normal(mean, cov_mat, 100).T
plt.figure(figsize=(4, 4))
plt.plot(x, y, 'x')
plt.axis([0,100, 1, 100])
text = "Covariance={:.3f}"
plt.title(text.format(cov));
```

```python
# 生成データの統計量算出
x_mean = np.mean(x)
y_mean = np.mean(y)
x_var = np.var(x)
y_var = np.var(y)
covar = np.sum((x-x_mean) * (y-y_mean))/len(x)
r_coef = covar / np.sqrt(x_var*y_var)

print("x平均: ", x_mean, "// y軸平均: ", y_mean)
print("x分散: ", x_var, "// y分散: ", y_var)
print("共分散: ", covar)
print("相関係数: ", r_coef)
```

```python
# 式(9.2)による共分散の算出
np.mean(x*y)-x_mean*y_mean
```
